```python
class book:
    def __init__(self, Name, Author, UserRating):
        self.Name = Name
        self.Author = Author
        self.UserRating = UserRating 
```

```python
nbabestsellers = []
with open("bestsellers.csv") as BooksData:
    reader = csv.DictReader(BooksData)
    for row in reader:
        Name = row['Name']
        Author = row['Author']
        UserRating = row['UserRating']
        nbabestsellers.append(book(Name, Author, UserRating))
```

```python
print(nbabestsellers[22].Name)
print(nbabestsellers[22].Author)
```

```python
bestsellersGraph = nx.Graph()
```

```python
bestsellersInfo = nbabestsellers[0:100]
```

```python
for f in bestsellersInfo:
    bestsellersGraph.add_node(f,Name = (f.Name[:15]))
```

```python
for f1 in bestsellersInfo:
    for f2 in bestsellersInfo:
        if not f1 == f2:
            if f1.Author == f2.Author:
                bestsellersGraph.add_edge(f1,f2)
```

```python
for f3 in bestsellersInfo:
    for f4 in bestsellersInfo:
        if not f3 == f4:
            if f3.UserRating == f4.UserRating:
                bestsellersGraph.add_edge(f3,f4)
```

```python
labels = nx.get_node_attributes(bestsellersGraph, 'Name')
print(labels)
```

```python
options = {'node_color':'chocolate', 'node_size':100, 'edge_color':'pink','width':3, 'labels':labels, 'with_labels': True}
plt.figure(1, figsize = (60, 60))
nx.draw(bestsellersGraph, **options)
```
