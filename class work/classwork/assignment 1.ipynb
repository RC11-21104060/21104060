### The Concept


Greedy algorithm: 

1.It is an algorithm that chooses the best or optimal (that is, the most advantageous) choice in the current state in each step of the selection, and hopes that the result is or optimal.

2.Successive choices are made in a top-down, iterative manner, reducing the problem to a smaller sub-problem with each greedy choice made.


<br/><br/><br/>



-----------------------------------------------------------------------------------------------------------------------------------------



### When is the greedy algorithm used?


1.When the current choice does not affect future choices, the problem can be solved optimally by the greedy algorithm.

2.The complexity of the original problem is so high that it is difficult to build a mathematical model of the optimal solution at once and the computational effort to find the optimal solution is too large.

3.When the optimal solution is not required, but can be solved by comparing optimal solutions.

##### Change issues


A cashier who needs to make change for $88. How to make change with the minimum number of notes/coins required.

Find the largest banknotes in order, e.g. to make change for ¥88, ¥88 = ¥50 + ¥20 + ¥10 + ¥5 + ¥1 * 3

```python
 Each bit of the arr: corresponds to 100, 50, 20, 10, 5 and 1 notes respectively.
arr = [0,0,0,0,0,0]

def change(money):
    while money >= 1:
        if money >= 100:
            money -= 100
            arr[0] += 1
        elif money >= 50:
            money -= 50
            arr[1] += 1
        elif money >= 20:
            money -= 20
            arr[2] += 1
        elif money >= 10:
            money -= 10
            arr[3] += 1
        elif money >= 5:
            money -= 5
            arr[4] += 1
        elif money >= 1:
            money -= 1
            arr[5] += 1

change(88)
print arr
```

One day, pirates intercept a cargo ship filled with a variety of antiques, each of which is worth a great deal of money.

Although the pirate ship is large enough to carry c, the weight of each antique is w How do the pirates load as many treasures as possible onto the pirate ship?


<br/><br/><br/>
_________________________________________________________________________________________________________________________________________




###### 1.Problem analysis

According to the problem description, this is an optimal loading problem that can be solved by a greedy algorithm. The number of items to be loaded is required to be as large as possible, and the capacity of the ship is fixed, so priority is given to items with a small weight, and the most items are loaded in the case of a fixed capacity. The greedy selection strategy of loading the lightest weight first is used to reach the global optimum from the local optimum, thus producing an optimal solution to the optimal loading problem.

<br/><br/>

###### 2.Algorithm design

(1) When the weight of the load is a fixed value c, the smaller the load, the greater the number of antiques n that can be loaded. Just choose the minimum weight antique in turn until it can no longer be loaded. 
<br/>

(2) The weight of the n antiques is sorted from smallest to largest (non-decreasing), and then the first i antiques are selected as many as possible according to the greedy strategy, until they cannot be loaded, at which point the optimum is reached.
